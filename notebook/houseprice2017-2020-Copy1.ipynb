```python
URL='https://www.businessinsider.com/home-value-home-price-change-in-50-years-every-state-2018-12'
response1 = requests.get(URL)
```

```python
title2017=soup1.findAll('h2')
title2017
```

```python
title2017[0].text
```

```python
state=[]
for i in title2017:
    state.append(i.text)
```

```python
states=pd.DataFrame(state)
states=states[0].str.split(".",expand=True)[:-1]
states=states.drop(columns={0})
states=states.rename(columns={1:'name'})
states
```

```python
cost=soup1.findAll('p')
cost
```

```python
sp=soup1.findAll('span')
sp
```

```python
prices=[]
price_spans = soup1.find_all('span', attrs={'data-sheets-value': True})

for span in price_spans:
    # Use regular expression to extract the number inside the 'data-sheets-value' attribute
    match = re.search(r'"3":(\d+)', span['data-sheets-value'])
    if match:
        price = match.group(1)
        prices.append(int(price))
prices
```

```python

houseprice_2017=[]
for i in range(1,52):
    houseprice_2017.append(prices[1+(i-1)*3])
hprice=pd.DataFrame(houseprice_2017)
hprice=hprice.rename(columns={0:'house_price'})
```

```python
states['house_price']=hprice['house_price']
states['year']=2017
df_2017=states
df_2017
```
