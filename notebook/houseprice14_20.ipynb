```python
import pandas as pd
```

```python
df=pd.read_csv('../data/all_st.csv')
df
```

```python
df_2018=df.filter(regex='^2018',axis=1)
df_2020=df.filter(regex='^2020',axis=1)
df_2016=df.filter(regex='^2016',axis=1)
df_2015=df.filter(regex='^2015',axis=1)
df_2014=df.filter(regex='^2014',axis=1)
```

```python
df_2014['avg']=df_2014.mean(axis=1)
df_2015['avg']=df_2015.mean(axis=1)
df_2016['avg']=df_2016.mean(axis=1)
df_2018['avg']=df_2018.mean(axis=1)
df_2020['avg']=df_2020.mean(axis=1)
df_2014['avg']=df_2014['avg'].fillna(0).astype('int')
df_2015['avg']=df_2015['avg'].fillna(0).astype('int')
df_2016['avg']=df_2016['avg'].fillna(0).astype('int')
df_2018['avg']=df_2018['avg'].fillna(0).astype('int')
df_2020['avg']=df_2020['avg'].fillna(0).astype('int')
```

```python
h=df[['RegionName']]
h=h.rename(columns={'RegionName':'name'})
```

```python
h_2020=h
h_2020['year']=2020
h_2020['house_price']=df_2020['avg']
h_2020
```
